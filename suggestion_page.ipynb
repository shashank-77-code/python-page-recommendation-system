```python
import json

# fun to load the json data 

def load_data (file_name):
    with open(file_name, "r") as f:
        return json.load(f)

#fun to perdict the users next page that he might like! 
def peges_you_may_like(user_id, data):
    pages_liked ={}
    #populate the user
    for user in data['users']:
        pages_liked[user['id']] = set(user['liked_pages'])

    if user_id not in pages_liked:
        return []
    pages_user_liked = pages_liked[user_id]

    page_suggestion = {}
    for other_users, pages in pages_liked.items():
        if other_users != user_id:
            shared_pages = pages_user_liked.intersection(pages)
        for page in pages:
            if page not in pages_user_liked:
                page_suggestion[page] = page_suggestion.get(page, 0) + len(shared_pages)  
    sorted_pages = sorted(page_suggestion.items(), key = lambda x : x[1], reverse = True)
    return [(page_id, score) for page_id, score in sorted_pages if score>= 1 ]

data = load_data("massive_data.json")
user_id = 2
page_recommendation = peges_you_may_like(user_id, data)
print(page_recommendation)
    
```
